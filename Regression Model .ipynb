```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn import linear_model
import warnings
%matplotlib inline
```

```python
df = pd.read_csv('crimestats.csv')

df['Population'] = df['Population'].str.replace(",","").astype(float)
df['Rape\n(legacy\ndefinition)2'] = df['Rape\n(legacy\ndefinition)2'].str.replace(",","").astype(float)
df['Violent\ncrime'] = df['Violent\ncrime'].str.replace(",","").astype(float)
df['Robbery'] = df['Robbery'].str.replace(",","").astype(float)
df['Aggravated\nassault'] = df['Aggravated\nassault'].str.replace(",","").astype(float)
df['Burglary'] = df['Burglary'].str.replace(",","").astype(float)
df['Property\ncrime'] = df['Property\ncrime'].str.replace(",","").astype(float)
df['Larceny-\ntheft'] = df['Larceny-\ntheft'].str.replace(",","").astype(float)
df['Motor\nvehicle\ntheft'] = df['Motor\nvehicle\ntheft'].str.replace(",","").astype(float)
df.dtypes
```

```python
df = df.fillna(0)
df['Murder']=df['Murder and\nnonnegligent\nmanslaughter']
df['Rape'] = df['Rape\n(revised\ndefinition)1'] +df['Rape\n(legacy\ndefinition)2']
df['Violent Crime']= df['Violent\ncrime']
df['Other Crime']= df['Robbery'] + df['Aggravated\nassault'] + df['Property\ncrime'] + df['Burglary']+ df['Larceny-\ntheft']+df['Motor\nvehicle\ntheft']+df['Arson3']
df['Property Crime']=df['Property\ncrime']

report = pd.DataFrame(df,
                      columns=['City','Population','Murder','Robbery','Rape','Violent Crime','Other Crime','Property Crime']
                     )
report
```

```python
report=report.drop([348,349,350])
report
```

```python
report.dtypes
```

```python
report['Murder']= np.where(report['Murder']>0,1,0)
```

```python
report['Robbery']= np.where(report['Robbery']>0,1,0)
```

```python
report['Population_squared'] = report['Population'] **2
```

```python
correlation_matrix = report.corr()
display(correlation_matrix)
```

```python
warnings.filterwarnings(
    action="ignore",
    module="scipy",
    message="^internal gelsd"
)
```

```python
%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

regr = linear_model.LinearRegression()
Y = report['Population'].values.reshape(-1, 1)
X = report[['Murder','Robbery','Rape','Violent Crime','Other Crime','Population_squared','Property Crime']]
regr.fit(X, Y)

print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))
```
